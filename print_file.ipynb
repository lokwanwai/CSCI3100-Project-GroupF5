```python
import os

def print_file_structure(start_path):
    for root, dirs, files in os.walk(start_path):
        level = root.replace(start_path, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print(f"{subindent}{f}")

# Specify the path to the repo you want to scan
repo_path = 'C:/Users/antho/Desktop/CSCI3100_Proj/CSCI3100-Project-GroupF5-main'
print_file_structure(repo_path)
```

```python
import os

def print_file_structure(start_path, exclude_dirs=None):
    exclude_dirs = set(exclude_dirs or [])
    for root, dirs, files in os.walk(start_path, topdown=True):
        dirs[:] = [d for d in dirs if d not in exclude_dirs]  # Exclude specified directories
        level = root.replace(start_path, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print(f"{subindent}{f}")

# Specify the path to the repo you want to scan
repo_path = 'C:/Users/antho/Desktop/CSCI3100_Proj/CSCI3100-Project-GroupF5/frontend'
exclude_directories = ['node_modules']  # Directories to exclude
print_file_structure(repo_path, exclude_directories)
```
